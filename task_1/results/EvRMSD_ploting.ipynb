```python
import pandas
from matplotlib import pyplot as plt
```

```python
prefix ="c18"
full_table = pandas.read_table("../analysis/%s/%s_scores.out" % (prefix, prefix), delim_whitespace=True, skiprows=1)
title = "%s docking: E v RMSD" % prefix
plt.scatter(full_table["ligand_rms_no_super_X"], full_table["interface_delta_X"], s=10, facecolors='none', edgecolors='b')
plt.xlabel("RMSD (A) to the native model")
plt.ylabel("Interface Energy (REU)")
plt.ylim(-24, 0)
plt.xlim(-1, 12)
plt.title(title)
plt.savefig("../analysis/%s_docking_EvRMSD.png" % prefix, bbox_inches="tight", dpi=300)
```
